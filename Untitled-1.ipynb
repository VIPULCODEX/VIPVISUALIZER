```python
words = ['artificial', 'transforms', 'industries']
weights = [2, 3, 1]
probs = [0.40, 0.35, 0.25]

hi, lo = probs.index(max(probs)), probs.index(min(probs))
semantic_scores = [w * p for w, p in zip(weights, probs)]
expectation = sum(semantic_scores) / len(words)

print("Probabilities:", dict(zip(words, probs)))
print("Highest:", words[hi])
print("Semantic scores:", {word: round(score, 4) for word, score in zip(words, semantic_scores)})
print("Expectation:", round(expectation, 4))
print("Spread:", round(max(probs) - min(probs), 4))
print("Most likely:", words[hi])
print("Least likely:", words[lo])
```

```python
import math

words = ['artificial', 'transforms', 'industries']
t_idx, c_idxs, weights = 1, [0, 2, 3], [2, 3, 1]

W = [[0.5, -0.2], [0.1, 0.3], [-0.3, 0.8], [0.4, -0.6]]
W_p = [[0.2, 0.1, -0.4, 0.3], [-0.1, 0.5, 0.2, -0.3]]
lr = 0.25

# --- Train for 2 Epochs ---
for epoch in range(2):
    h = W[t_idx]
    u = [sum(W_p[a][k] * h[a] for a in range(2)) for k in range(4)]
    exp_u = [math.exp(x) for x in u]
    probs = [ex / sum(exp_u) for ex in exp_u]
    loss = -sum(math.log(probs[c]) for c in c_idxs)
    
    print(f"Epoch {epoch} | Loss: {loss:.4f} | Probs: {[round(probs[c], 4) for c in c_idxs]}")
    
    grad_u = [3 * probs[k] - (1 if k in c_idxs else 0) for k in range(4)]
    grad_h = [sum(W_p[a][k] * grad_u[k] for k in range(4)) for a in range(2)]
    
    for a in range(2):
        W[t_idx][a] -= lr * grad_h[a]
        for k in range(4):
            W_p[a][k] -= lr * grad_u[k] * h[a]

# --- Final Results ---
c_probs = [probs[c] for c in c_idxs]
hi, lo = c_probs.index(max(c_probs)), c_probs.index(min(c_probs))
semantic_scores = [w * p for w, p in zip(weights, c_probs)]

print("\n--- FINAL TRAINING RESULTS ---")
print("Probabilities:", dict(zip(words, [round(p, 4) for p in c_probs])))
print("Semantic Scores:", dict(zip(words, [round(s, 4) for s in semantic_scores])))
print("Expectation:", round(sum(semantic_scores) / 3, 4))
print("Spread:", round(max(c_probs) - min(c_probs), 4))
print("Most likely:", words[hi])
print("Least likely:", words[lo])
```
